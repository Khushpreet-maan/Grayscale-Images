## **Resize the 100 images to 50% (smaller)" using OpenCV**
---

```python
import cv2
import os
from bing_image_downloader import downloader
```

```python
# Step 1: Download 100 dog images (auto creates folder)
query = "dog"
download_dir = "input_images"
downloader.download(query, limit=100, output_dir=download_dir, adult_filter_off=True, force_replace=False, timeout=60)

# Define input and output folders
input_folder = os.path.join(download_dir, query)
output_folder = "output_resized_grayscale_images"
os.makedirs(output_folder, exist_ok=True)
```

```python
# Step 2: Process the images
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]

for i, image_name in enumerate(image_files):
    img_path = os.path.join(input_folder, image_name)
    image = cv2.imread(img_path)

    if image is None:
        print(f"Skipping: {img_path}")
        continue

    # Resize image to 50%
    height, width = image.shape[:2]
    resized_image = cv2.resize(image, (width // 2, height // 2))

    # Convert to grayscale
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Save the processed image
    output_path = os.path.join(output_folder, f"resized_gray_{image_name}")
    cv2.imwrite(output_path, gray_image)

    print(f"[{i+1}] Processed and saved: {output_path}")
    print("✅ Done: All 100 images downloaded, resized to 50%, and converted to grayscale.")
```
