## **"Convert 100 images to Grey Scale (b/w)" using OpenCV**

```python
import cv2
import os
from bing_image_downloader import downloader
```

**Step 1: Download 100 dog images**

```python
query = "dog"
download_dir = "input_images"
downloader.download(query, limit=100, output_dir=download_dir, adult_filter_off=True, force_replace=False, timeout=60)

# Find the folder where images are saved
input_folder = os.path.join(download_dir, query)
output_folder = "output_grayscale_images"
```

**Step 2: Create output directory**

```python
os.makedirs(output_folder, exist_ok=True)
```

**Step 3: Convert all downloaded images to grayscale**

```python
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]

for i, image_name in enumerate(image_files):
    img_path = os.path.join(input_folder, image_name)
    image = cv2.imread(img_path)

    if image is None:
        print(f"Skipping: {img_path} (not a valid image)")
        continue

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Save the result
    output_path = os.path.join(output_folder, f"gray_{image_name}")
    cv2.imwrite(output_path, gray)

    print(f"[{i+1}] Converted: {image_name}")

print("✅ All 100 images downloaded and converted to grayscale.")
```

**Set the input and output directory**

```python
input_folder = "input_images"
output_folder = "output_grayscale_images"
```

```python
# Create output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get list of all image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

# Convert first 100 images to grayscale
for i, image_name in enumerate(image_files[:100]):
    img_path = os.path.join(input_folder, image_name)
    image = cv2.imread(img_path)

    if image is None:
        print(f"Skipping: {img_path} (not a valid image)")
        continue

    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Save the grayscale image
    output_path = os.path.join(output_folder, f"gray_{image_name}")
    cv2.imwrite(output_path, gray_image)

    print(f"Converted {i+1}: {image_name} -> {output_path}")

print("✅ Conversion completed for up to 100 images.")
```
